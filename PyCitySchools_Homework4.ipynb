```python
import pandas as pd
```

```python
#students_complete file
file_one = "../../../working/04-Pandas/Homework/Instructions/PyCitySchools/Resources/students_complete.csv"
```

```python
#schools_complete file
file_two = "../../../working/04-Pandas/Homework/Instructions/PyCitySchools/Resources/schools_complete.csv"
```

```python
file_one_df = pd.read_csv(file_one)
```

```python
file_two_df = pd.read_csv(file_two)
```

```python
#Student Complete DataFrame
file_one_df.head()
```

```python
#Schools Complete DataFrame
file_two_df.head()
```

```python
#Merging two files together
complete_school_data_df = pd.merge(file_one_df, file_two_df, on = "school_name")
complete_school_data_df
```

# District Summary

```python
#Calculating total number of students
student_count = complete_school_data_df["Student ID"].count()
student_count
```

```python
#Calculating total number of schools
school_count = complete_school_data_df["school_name"].nunique()
school_count
```

```python
#Calculating total budget across all schools
total_budget = file_two_df["budget"].sum()
total_budget
```

```python
#Calculating average math scores for the district(all the schools)
average_math = complete_school_data_df["math_score"].mean()
average_math
```

```python
#Calculating average reading scores for the district
average_reading = complete_school_data_df["reading_score"].mean()
average_reading
```

```python
#Conditional statement for if a student is passing math or not
passing_math = complete_school_data_df["math_score"].loc[complete_school_data_df["math_score"] >= 70]

#Counting the number of students who are passing math
passing_math_count = passing_math.count()

#Finding the percent of students who are passing math
percent_passing_math = (passing_math_count / student_count) * 100
percent_passing_math
```

```python
#Conditional statement for if a student is passing reading or not
passing_reading = complete_school_data_df["reading_score"].loc[complete_school_data_df["reading_score"] >= 70]

#Counting the number of the students who are passing reading
passing_reading_count = passing_reading.count()

#Finding the percent of students who are passing reading
percent_passing_reading = (passing_reading_count / student_count) * 100
percent_passing_reading
```

```python
#Conditional statement for if a student is passing both math and reading
passing_both = complete_school_data_df["type"].loc[(complete_school_data_df["math_score"] >= 70) & (complete_school_data_df["reading_score"] >= 70)]

#Counting the number of students passing both math and reading
passing_both_count = passing_both.count()

#Finding the percent of students passing both math and reading
percent_passing_both = (passing_both_count / student_count) * 100
percent_passing_both
```

```python
#Creating a Summary Table for the District 
district_summary_df = pd.DataFrame(
    {"Total Schools": [school_count],
     "Total Students": [student_count],
     "Total Budget": [total_budget],
     "Average Math Score": [average_math],
     "Average Reading Score": [average_reading],
     "% Passing Math": [percent_passing_math],
     "% Passing Reading": [percent_passing_reading],
     "%Overall Passing": [percent_passing_both]
    }
)

district_summary_df
```

# School Summary

```python
#Determining school type for each school
school_type = file_two_df.set_index(["school_name"])["type"]
school_type
```

```python
#Calculating number of students per school
grouped_school_sum_df = complete_school_data_df["school_name"].value_counts()                                              
grouped_school_sum_df
```

```python
#Calculating the total budget for each school
grouped_school_sum_df2 = file_two_df.groupby(["school_name"]).sum()['budget']

grouped_school_sum_df2.head()
```

```python
#Calculating Per Student Budget
per_stu_budget = (grouped_school_sum_df2) / (grouped_school_sum_df)
per_stu_budget
```

```python
#Calculating Percent Passing Math

#Calculating how many students are passing math in each school
num_passing_math = complete_school_data_df[(complete_school_data_df["math_score"] >= 70)]
num_passing_math = num_passing_math.groupby(["school_name"]).count()["student_name"]


#Calculating percent of students in each school who are passing
percent_passing_math_school = (num_passing_math / grouped_school_sum_df) * 100
percent_passing_math_school
```

```python
#Calculating Percent Passing Reading

#Calculating how many students are passing reading in each school
num_passing_reading = complete_school_data_df[(complete_school_data_df["reading_score"] >= 70)]
num_passing_reading = num_passing_reading.groupby(["school_name"]).count()["student_name"]
num_passing_reading.head()

#Calculating percent of students in each school who are passing
percent_passing_reading_school = (num_passing_reading / grouped_school_sum_df) * 100
percent_passing_reading_school
```

```python
#Calculating Percent Passing Overall

#Calculating how many students are passing overall in each school
num_passing_overall = complete_school_data_df[(complete_school_data_df["reading_score"] >= 70) & (complete_school_data_df["math_score"] >= 70)]
num_passing_overall = num_passing_overall.groupby(["school_name"]).count()["student_name"]

#Calculating percent of students in each school who are passing
percent_passing_overall_school = (num_passing_overall / grouped_school_sum_df) * 100
percent_passing_overall_school
```

```python
#Calculating the average reading and math scores per school
avg_math_grouped = complete_school_data_df.groupby(["school_name"]).mean()["math_score"]
avg_reading_grouped = complete_school_data_df.groupby(["school_name"]).mean()["reading_score"]
```

```python
#Creating the School Summary table
school_sum_df = pd.DataFrame(
    {"Type": school_type,
     "Total Students": grouped_school_sum_df,
     "Total School Budget": grouped_school_sum_df2,
     "Per Student Budget": per_stu_budget,
     "Average Math Score": avg_math_grouped,
     "Average Reading Score": avg_reading_grouped,
     "% Passing Math" : percent_passing_math_school,
     "% Passing Reading" : percent_passing_reading_school,
     "% Overall Passing" : percent_passing_overall_school
    }

)

school_sum_df.head()
```

# Top Performing Schools (By % Overall Passing)

```python
#Sorting School Summary table (highest to lowest)
sorting_top_schools = school_sum_df.sort_values(by = "% Overall Passing", ascending = False)
sorting_top_schools.head()
```

# Bottom Performing Schools (By % Overall Passing)

```python
#Sorting School Summary table (lowest to highest)
sorting_bottom_schools = school_sum_df.sort_values(by = "% Overall Passing")
sorting_bottom_schools.head()
```

# Math Scores by Grade

```python
#Conditional statment for if a student is in the 9th grade
nineth_grade_df = complete_school_data_df.loc[complete_school_data_df["grade"] == "9th"]

#For 9th graders, this is grouping them by school
nineth_school_df = nineth_grade_df.groupby(["school_name"])
nineth_school_df.count().head()
```

```python
#Calculating the average math score for 9th graders in a given school
nineth_mean_math = nineth_school_df["math_score"].mean()

#Calculating the average reading score for 9th graders in a given school
nineth_mean_reading = nineth_school_df["reading_score"].mean()
```

```python
#Conditional statment for is a student is in 10th grade
tenth_grade_df = complete_school_data_df.loc[complete_school_data_df["grade"] == "10th"]

#Grouping the 10th grade students by school
tenth_school_df = tenth_grade_df.groupby(["school_name"])

#Calculating the average of 10th graders' math scores for each school
tenth_mean_math = tenth_school_df["math_score"].mean()

#Calculating the average reading score for 10th graders in a given school
tenth_mean_reading = tenth_school_df["reading_score"].mean()
tenth_mean_reading
```

```python
#Conditional statment for is a student is in 11th grade
eleventh_grade_df = complete_school_data_df.loc[complete_school_data_df["grade"] == "11th"]

#Grouping the 11th grade students by school
eleventh_school_df = eleventh_grade_df.groupby(["school_name"])

#Calculating the average of 11th graders' math scores for each school
eleventh_mean_math = eleventh_school_df["math_score"].mean()

#Calculating the average reading score for 11th graders in a given school
eleventh_mean_reading = eleventh_school_df["reading_score"].mean()
#eleventh_mean_reading
```

```python
#Conditional statment for is a student is in 12th grade
twelfth_grade_df = complete_school_data_df.loc[complete_school_data_df["grade"] == "12th"]

#Grouping the 12th grade students by school
twelfth_school_df = twelfth_grade_df.groupby(["school_name"])

#Calculating the average of 12th graders' math scores for each school
twelfth_mean_math = twelfth_school_df["math_score"].mean()

#Calculating the average reading score for 12th graders in a given school
twelfth_mean_reading = twelfth_school_df["reading_score"].mean()
#twelfth_mean_reading
```

```python
#Merging 9th and 10th graders math scores 
math_merge_df1 = pd.merge(nineth_mean_math, tenth_mean_math, on = "school_name")

#Renaming columns in chart to 9th and 10th
renamed_math_merge_1 = math_merge_df1.rename(columns={"math_score_x": "9th", "math_score_y": "10th"})
renamed_math_merge_1.head()
```

```python
#Merging both charts together to get grades 9th-12th
math_merge_df3 = pd.merge(renamed_math_merge_1, renamed_math_merge_2, on = "school_name")

math_merge_df3.head()

```

Trends Based on this DataSet:
1. Charter schools have higher passing rates than district schools. When we sorted our School Summary table from the school with the highest overall passing rate to the lowest, the top 5 schools were Charter. When we sorted the schools from the lowest overall passing rate to the higest, the top 5 schools were District. We can also see this in our Scores by School Type table. The scores for charter schools were higher in every category. 
2. There is an inverse relationship between the budget of the school and student success; more money does not mean better results. You can see this by looking at the "Per Student Budget" column in the tables that sorted the School Summary table by the highest overall passing percentage and the lowest overall passing percentange. You can also see this in the Scores by School Spending table.
